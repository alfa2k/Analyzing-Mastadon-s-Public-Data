```python
import time

user_ids = set()
max_id = None
desired_count = 500

while len(user_ids) < desired_count:
    public_toots = mastodon.timeline_public(limit=40, max_id=max_id)
    if not public_toots:
        break

    for toot in public_toots:
        user = toot['account']
        user_id = user['id']
        user_ids.add(user_id)

    max_id = public_toots[-1]['id']

    time.sleep(1)

print(f"Collected {len(user_ids)} user IDs.")
```

```python
from mastodon import Mastodon
import json

# initializing Mastodon API
mastodon = Mastodon(
    access_token='pytooter_usercred.secret',
    api_base_url='https://mastodon.social'
)

# func to fetch user data by user ID
def fetch_user_data(user_id):
    user = mastodon.account(user_id)
    user['created_at'] = user['created_at'].isoformat()
    user['last_status_at'] = user['last_status_at'].isoformat() if 'last_status_at' in user else None
    # Fetch followers and following
    user['followers'] = fetch_followers(user_id)
    user['following'] = fetch_following(user_id)
    return user

# getting followers for a user
def fetch_followers(user_id):
    followers = mastodon.account_followers(user_id)
    return [follower['id'] for follower in followers]

# getting accounts a user is following
def fetch_following(user_id):
    following = mastodon.account_following(user_id)
    return [followed['id'] for followed in following]

seed_user_ids = first_250_user_ids
user_data = []
for user_id in seed_user_ids:
    try:
        user = fetch_user_data(user_id)
        user_data.append(user)
    except Exception as e:
        print(f"Error fetching data for user {user_id}: {e}")

# saving the collected data to a JSON file
with open('user_data.json', 'w') as file:
    json.dump(user_data, file, ensure_ascii=False, indent=4)

# the file is saved in drive
```
